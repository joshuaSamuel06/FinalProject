```python
#creaing a function to seperate the unique names from all the rows into a list
def seperate(all_list):
    listing = []
    for l in all_list:
        if isinstance(l,str):
            a=l.split(";")
            for j in a:
                if j not in listing and isinstance(j,str):
                    listing.append(j)
    return listing
```

```python
#creating a dictionery of programming languages that people know with their counts 
Alllanguages = Data["LanguageHaveWorkedWith"].tolist()

languages= seperate(Alllanguages)
LangDict = {}
for i in languages:
    LangDict[i] = Data["LanguageHaveWorkedWith"].str.contains(i, na=False).sum()

#creating a dictionery of programming languagesthat people want to learn or work with and the number of instances
Alllanguages_want = Data["LanguageWantToWorkWith"].tolist()

languages_want = seperate(Alllanguages_want)
LangDict_want = {}
for i in languages_want:
    LangDict_want[i] = Data["LanguageWantToWorkWith"].str.contains(i, na=False).sum()
```

### Skills Analysis

```python
cols = ["Age", "EdLevel", "Country", "LanguageHaveWorkedWith", 'PlatformHaveWorkedWith',
        'WorkExp', 'OrgSize', 'Industry', 'ToolsTechHaveWorkedWith', 'ConvertedCompYearly']     # Listing the needed columns to keep
data = main_data.loc[:, cols]
data.rename(columns={"LanguageHaveWorkedWith":"ProgrammingLanguages",
                     "PlatformHaveWorkedWith":"UsedPlatform",
                     "ToolsTechHaveWorkedWith":"UsedTools",
                     "ConvertedCompYearly":"AnnualSalary($)"}, inplace=True)
data["Age"] = data["Age"].str.replace("years old", "")
data.dropna(subset=["ProgrammingLanguages", "AnnualSalary($)"], inplace=True)   # Dropping empty cells
data.head(2)
```

```python
# Extracting all skills found in the dataset
def flatten(xss):
    return [x for xs in xss for x in xs]

skills_lists = set(data['ProgrammingLanguages'])
languages = []
for i in skills_lists:
    if type(i) == str:
        languages.append(i.split(';'))

skills = list(set(flatten(languages)))

for index, skill in enumerate(skills):
    if len(skill) > 1:
        stop = skill.find("(")
        if stop > 0:
            skills[index] = skill[0:stop].strip()
        if skill == 'C++':
            skills[index] = 'C+'
```

```python
skills_entries = []
for i in skills:
    skills_entries.append((i, len(data[(data['ProgrammingLanguages'].str.contains(i, na=False))])))

print(skills_entries)
```

```python
sorted_skills = sorted(skills_entries, key=lambda x: x[1])
sorted_skills
# Skills data analysis
y = [(v[1]/data.shape[0])*100 for i, v in enumerate(sorted_skills[-11:-1])]
x = [v[0] for i, v in enumerate(sorted_skills[-11:-1])]
x[-1] = 'C/ C++'
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
colors = cm.rainbow(np.linspace(0.3, 1, len(y)))
fig = plt.figure()
plt.bar(x, y)
plt.xticks(rotation=30)
plt.xlabel("Skill")
plt.ylabel("Ratio of people (%)")
plt.title("Top 10 used skills")
fig.savefig('Top_skills(vis1).png', bbox_inches = 'tight', dpi=720)
plt.show()
```

```python
import re
db_worked_with = list(main_data['DatabaseHaveWorkedWith'].dropna())
dbs = {}

def add_to_dict(a, dt):
    if a in dt:
        dt[a] += 1
    else:
        dt[a] = 1
    return dt

for i in db_worked_with:
    if ';':
        x = i.split(';')
        for j in x:
            dbs = add_to_dict(j, dbs)
    else:
        dbs = add_to_dict(i, dbs)

dbs2 = {}
env_list = []
ratios = []
count = 0
for k, v in dbs.items():
    count += v

for k, v in dbs.items():
    value = round(v / count * 100, 2)
    if value > 2.5:
        env_list.append(k)
        ratios.append(value)

dbs2['Database Environment'] = env_list
dbs2['Ratio'] = ratios

print(dbs2)
```

```python
import plotly.express as px
fig = px.pie(dbs2, values='Ratio', names='Database Environment', title='Top Utilized Database Environments (%)')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=15,
                  marker=dict(colors=colors, line=dict(color='#000000', width=1)))
fig.write_html('./Database_environments(vis4).html')
fig.write_image("Database_environments(vis4).png", format = 'png', width = 1000, height = 500, scale = 2, engine = 'kaleido')
fig.show()
```

```python
# Analyze preference variation based on different factors : Industry
factor = 'Industry'

preference_variation = df.groupby(['RemoteWork', factor]).size().unstack()

total_responses = preference_variation.sum(axis=1)

percentage_variation = preference_variation.divide(total_responses, axis=0) * 100

for remote_pref in percentage_variation.index:
    fig, ax = plt.subplots(figsize=(8, 9))
    
    colors = plt.cm.tab20c.colors if remote_pref == 'Remote' else plt.cm.tab20b.colors
    ax.pie(percentage_variation.loc[remote_pref], labels=percentage_variation.columns, autopct='%1.1f%%', colors=colors, startangle=90)
    ax.set_title(f'{remote_pref} Work Preference', fontsize=14)
    ax.axis('equal')  
    ax.legend(loc='upper left', bbox_to_anchor=(1, 0.5))
    
    plt.show()
```

Interquartile Range (IQR): The box itself spans the IQR, which contains the middle 50% of the data.
Whiskers: Lines extending from the box represent the range of the data, excluding outliers. They typically extend 1.5 times the IQR from the first and third quartiles.
Outliers: Individual data points lying beyond the whiskers are considered outliers and are shown separately.

```python

tbranch_respondents = df[df['TBranch'].notna()]

age_groups = ['Under 18 years old', '18-24 years old', '25-34 years old', 
              '35-44 years old', '45-54 years old', '55-64 years old', 
              '65 years or older', 'Prefer not to say']

age_counts = tbranch_respondents['Age'].value_counts()

yes_counts = tbranch_respondents[tbranch_respondents['TBranch'] == 'Yes']['Age'].value_counts()
no_counts = tbranch_respondents[tbranch_respondents['TBranch'] == 'No']['Age'].value_counts()

plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(age_groups))
bar1 = plt.bar(index, yes_counts, bar_width, label='Yes', color='skyblue')
bar2 = plt.bar([i + bar_width for i in index], no_counts, bar_width, label='No', color='orange')

plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.title('Opinions on Participating in Professional Developer Series by Age Group')
plt.xticks([i + bar_width/2 for i in index], age_groups, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
